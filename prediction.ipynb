**위/경도로 변환(kakao map API)**

```python
import requests, json
import pandas as pd
import time
 
def get_location(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    headers = {"Authorization": "KakaoAK 5cb2bf7707aaa00826718af02f35373b"}
    api_json = json.loads(str(requests.get(url,headers=headers).text))
    return api_json

def result_location(i):
    api_json = get_location(token['byunhwan'][i])
    if api_json['documents']:
        address = api_json['documents'][0]['address']
        token.loc[i,'lon'] = address['x']; 
        token.loc[i,'lat'] = address['y']        
    else:
        token.loc[i,'lon'], token.loc[i,'lat'] = None, None
    print(i, '번째 위경도 변환')

i = 0
while i<=len(token['byunhwan']):
    try:
        result_location(i)
        i+=1
    except:
        break
```

```python
#df 폼맞춰주기
token2 = token[token['lon'].notnull()] #위경도 빈값 제거 -> 필요한것만!
token2['lon'] = token2['lon'].astype(float) # astype float
token2['lat'] = token2['lat'].astype(float) # astype float
token2['lon'] = np.trunc(token2['lon']*1000)/1000 # 소수점 3자리
token2['lat'] = np.trunc(token2['lat']*1000)/1000 # 소수점 3자리
token2['lonlat'] = token2['lon'].apply(str) + " " + token2['lat'].apply(str)
```
